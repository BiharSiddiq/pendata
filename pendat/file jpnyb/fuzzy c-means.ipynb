```python
# üßÆ Import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# üîß Fuzzy C-Means Algorithm
def fuzzy_c_means(X, n_clusters=2, m=2, max_iter=100, error=1e-5):
    n_samples = X.shape[0]
    U = np.random.dirichlet(np.ones(n_clusters), size=n_samples)

    for iteration in range(max_iter):
        U_old = U.copy()
        centers = np.dot(U.T ** m, X) / np.sum(U.T ** m, axis=1, keepdims=True)

        for i in range(n_samples):
            for j in range(n_clusters):
                denominator = sum(
                    (np.linalg.norm(X[i] - centers[j]) / np.linalg.norm(X[i] - centers[k])) ** (2 / (m - 1))
                    for k in range(n_clusters)
                )
                U[i, j] = 1 / denominator

        if np.linalg.norm(U - U_old) < error:
            break

    return centers, U
```

```python
# üìä Membaca file dan menjalankan FCM
# Ganti nama file jika berbeda
file_name = next(iter(uploaded))  # otomatis ambil nama file pertama
df = pd.read_excel(file_name, sheet_name='Lembar1')

# Normalisasi data
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(df)

# Jalankan Fuzzy C-Means (default: 2 klaster)
centers, membership = fuzzy_c_means(data_scaled, n_clusters=2)

# Gabungkan hasil ke DataFrame
membership_df = pd.DataFrame(membership, columns=[f'Cluster {i+1}' for i in range(membership.shape[1])])
result_df = pd.concat([df, membership_df], axis=1)

# Simpan hasil ke Excel
output_file = "hasil_fuzzy_c_means.xlsx"
result_df.to_excel(output_file, index=False)

# Tampilkan beberapa hasil
result_df.head()
```
